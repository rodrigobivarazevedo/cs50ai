```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
from sklearn.model_selection import train_test_split


# save filepath to variable for easier access
melbourne_file_path = 'melb_data.csv'
# read the data and store data in DataFrame titled melbourne_data
melbourne_data = pd.read_csv(melbourne_file_path)
```

```python
melbourne_data = melbourne_data.dropna(axis=0)
melbourne_data.info()
```

```python
melbourne_features = ['Rooms', 'Type', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude',
       'Longtitude', 'Propertycount']
```

```python
# select data corresponding to features in melbourne_features
X = melbourne_data[melbourne_features]
# select data corresponding to the Price
y = melbourne_data.Price
```

```python
# one-hot encoding via pandas

pandas_encoded = pd.get_dummies(X, dtype=int)
```

```python
import pandas as pd
from sklearn.preprocessing import OneHotEncoder

# Separate the column to encode
X_to_encode = X[['Type']]

# Initialize the OneHotEncoder
onehot_encoder = OneHotEncoder()

# Fit and transform the column to encode
type_encoded = onehot_encoder.fit_transform(X_to_encode)

# Convert the sparse matrix to a DataFrame
encoded_df = pd.DataFrame.sparse.from_spmatrix(type_encoded, columns=onehot_encoder.categories_[0])

# Concatenate the original DataFrame and the encoded DataFrame
X_encoded = pd.concat([X.drop(columns=['Type']), encoded_df], axis=1)

# Print the shapes to verify
print("Original X shape:", X.shape)
print("Encoded X shape:", X_encoded.shape)
print("Encoded X shape:", X_encoded.info())
```

```python
# sequential feature selection

from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split

feat_labels = ['Rooms', "Type", 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude',
       'Longtitude', 'Propertycount']

forest = RandomForestClassifier(n_estimators=500, random_state=1)
# Split data into training and validation sets
X_train, X_val, y_train, y_val = train_test_split(pandas_encoded, y, random_state=0)

forest.fit(X_train, y_train)
importances = forest.feature_importances_

indices = np.argsort(importances)[::-1]

for f in range(X_train.shape[1]):
    print("%2d) %-*s %f" % (f + 1, 30,
                            feat_labels[indices[f]],
                            importances[indices[f]]))

plt.title('Feature Importance')
plt.bar(range(X_train.shape[1]),
        importances[indices],
        align='center')

plt.xticks(range(X_train.shape[1]),
           feat_labels[indices], rotation=90)
plt.xlim([-1, X_train.shape[1]])
plt.tight_layout()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error

# Assuming X and y are already defined and contain your features and target variable
melbourne_features = ['Rooms', 'Distance', 'Postcode', 'Bedroom2', 'Bathroom', 'Car',
       'Landsize', 'BuildingArea', 'YearBuilt', 'Lattitude',
       'Longtitude', 'Propertycount']
# select data corresponding to features in melbourne_features
X = melbourne_data[melbourne_features]
y = melbourne_data.Price

# Split data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(X, y, random_state=0)

# Define the Random Forest model
forest_model = RandomForestRegressor(
    n_estimators=1000,   # Number of trees in the forest
    criterion='squared_error',    # Criterion for splitting: Mean Squared Error
    random_state=1,     # Random seed for reproducibility
    n_jobs=-1           # Use all available CPU cores
)

# Train the model on the training data
forest_model.fit(train_X, train_y)

# Make predictions on the validation data
val_predictions = forest_model.predict(val_X)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(val_y, val_predictions)
print("Validation MAE for Random Forest Model: {}".format(mae))
```

```python
from sklearn.model_selection import train_test_split, RandomizedSearchCV
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error
import numpy as np

# Assuming pandas_encoded and y are your features and target variables

# Split data into training and validation sets
train_X, val_X, train_y, val_y = train_test_split(pandas_encoded, y, random_state=0)

# Define the Random Forest model
forest_model = RandomForestRegressor(random_state=1, n_jobs=-1)

# Define the parameter grid to search through
param_grid = {
    'n_estimators': [100, 200, 300, 400, 500, 600, 700],  # Number of trees in the forest
    'max_depth': [None, 10, 50, 100, 200, 500, 600, 700],  # Maximum depth of each tree
    'min_samples_split': [2, 5, 10],  # Minimum number of samples required to split an internal node
    'min_samples_leaf': [1, 2, 4],  # Minimum number of samples required to be at a leaf node
    'max_features': ['auto', 'sqrt', 'log2'],  # Number of features to consider at each split
    'bootstrap': [True, False]  # Whether bootstrap samples are used when building trees
}

# Perform RandomizedSearchCV to find the best parameters
rf_random = RandomizedSearchCV(estimator=forest_model, param_distributions=param_grid,
                               n_iter=100, cv=3, verbose=2, random_state=42, n_jobs=-1)

# Fit the RandomizedSearchCV instance
rf_random.fit(train_X, train_y)

# Get the best parameters found by RandomizedSearchCV
print("Best parameters found by RandomizedSearchCV:")
print(rf_random.best_params_)

# Make predictions on the validation data using the best model found
val_predictions = rf_random.best_estimator_.predict(val_X)

# Calculate and print the Mean Absolute Error (MAE)
mae = mean_absolute_error(val_y, val_predictions)
print("\nValidation MAE for Random Forest Model after hyperparameter tuning: {:.2f}".format(mae))
```
